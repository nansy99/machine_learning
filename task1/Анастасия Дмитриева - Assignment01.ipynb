####    10. Постройте распределения расхода топлива по городу (city mpg) для автомобилей с ручной коробкой передач и автоматической. 
Не забудьте про подписи осей, размеры шрифтов, легенду.

```python
from matplotlib import pyplot
pyplot.hist(car[car['Transmission Type']=='AUTOMATIC']['city mpg'],bins=50, color = 'skyblue', alpha = 1, label = 'AUTOMATIC')
pyplot.hist(car[car['Transmission Type']=='MANUAL']['city mpg'],bins=30,color = 'plum', alpha = 0.7, label = 'MANUAL')
pyplot.legend(loc='upper right')
plt.xlabel('city mpg',fontsize=14)
plt.ylabel('frequency',fontsize=14)
plt.title('City mpg distribution')
pyplot.show()
plt.savefig("Dmitrieva_fig10.png")
```

####    11. Постройте зависимость средней стоимости автомобиля от числа цилиндров. 
Используйте функцию plt.errorbar, чтобы помимо средних значений отобразить среднеквадратичное отклонение величины.

```python
car_m=car.groupby('Engine Cylinders').mean()
car_std=car.groupby('Engine Cylinders').std()
plt.errorbar(x=car_m.index, y=car_m.MSRP, yerr=car_std.MSRP)
plt.xlabel('Engine Cylinders',fontsize=14)
plt.ylabel('Mean price',fontsize=14)
plt.title('Price from number of engine cylinders')
plt.savefig("Dmitrieva_fig11.png")
```

####    12. Постройте boxplot, отображающий распределение стоимости автомобилей следующих марок: Audi, Toyota, BMW, Mazda, Hyundai. 
Не забудьте про подписи осей, размеры шрифтов.

```python
msrp_audi=car[(car['Make'] == 'Audi')]['MSRP'].dropna().values
msrp_toyota=car[(car['Make'] == 'Toyota')]['MSRP'].dropna().values
msrp_bmw=car[(car['Make'] == 'BMW')]['MSRP'].dropna().values
msrp_mazda=car[(car['Make'] == 'Mazda')]['MSRP'].dropna().values
msrp_hyundau=car[(car['Make'] == 'Hyundai')]['MSRP'].dropna().values
```

```python
fig=plt.boxplot([msrp_audi,msrp_toyota,msrp_bmw,msrp_mazda,msrp_hyundau],labels=['Audi','Toyota','BMW','Mazda','Hyundai'])
plt.tick_params(axis='both', which='major', labelsize=14)
plt.ylabel('MSRP', fontsize=14)
plt.xlabel('Make', fontsize=14)
plt.savefig("Dmitrieva_fig12.png")
```
